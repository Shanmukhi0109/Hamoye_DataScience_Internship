```python
#Qs11 total sum of animal fat in year 2014,2017
df.groupby('Item')[['Y2014','Y2017']].sum()
```

```python
#Qs12 mean & standard deviation in 2016 year
mean=df['Y2015'].mean()
std=round(df['Y2015'].std(),3)
print(mean,std)
```

```python
#Qs13 total no & % of missing data in Y2016
total=df['Y2016'].isnull().sum()
percent=round((df['Y2016'].isnull().sum()*100)/len(df.index),2)
print(total,percent)
```
